## Problem 3: Gibbs phenomenon

### (a) Calculating the overshoot
For the truncated Fourier series for the square wave you analyzed in 1.b, you may notice that the Fourier series approximation $f_N(x)$ including the first $N$ non-zero terms always overshoots the exact result in the vicinity of the discontinuity at $x = 0$. We will define as $$ \text{overshoot} = \max f_N(x) - 1.$$ Because $f_N(x)$ is periodic, the maximum may be taken over a single period.

For example, the overshoot for $f_2$, including only the first two non-zero terms, may be calculated numerically as:

````python
x = np.linspace(0, 2*np.pi,num=1000)
f2 = 4/np.pi * np.sin(x) + 4/(3*np.pi) * np.sin(3 * x)

overshoot = np.amax(f2) - 1
print("overshoot =",overshoot)
````

And so we find that $f_2(x)$ overshoots one by about $0.2$.

### (b) Calculating the RMS error

Now, instead of the overshoot, consider the root-mean-square (RMS) error of the approximation for each value of $N$. The RMS error is defined as $$\sigma_N = \sqrt{\langle (f_N(x) - f_{sq}(x))^2\rangle},$$ where because this is a periodic function the mean $\langle (f_N(x) - f_{sq}(x))^2\rangle$ can be taken over one period. Because it's integrated over $x$, the RMS error can be thought of as the total difference between the two curves. For example, the RMS error for $f_2$ is can be calculated numerically as:

````python
x = np.linspace(0, 2*np.pi,num=1000)                   # define an array with 1000 points along the x-axis

fexact = np.sign(np.sin(x))                            # calculate the exact square wave
f2 = 4/np.pi * np.sin(x) + 4/(3*np.pi) * np.sin(3 * x) # calculate the Fourier series with two terms


sigma2 = np.sqrt(np.mean((f2 - fexact)**2))            # calculate the RMS error for f2
print('sigma2 =',sigma2)
````
